```python
address = 'Jl. Kol. Yos Sudarso No.8, Silalas, Kec. Medan Bar., Kota Medan, Sumatera Utara 20234, Indonesia'
```

```python
load_dotenv()

GEOCODING_API_KEY = os.getenv('GEOCODING_API_KEY')
```

```python
def get_lat_lng(address: str):
    r = requests.get(f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={GEOCODING_API_KEY}')
    results = r.json()['results'][0]
    return results['formatted_address'], results['geometry']['location']
```

```python
get_lat_lng(address)
```

```python
addresses: list[str] = []
lats: list[float] = []
lngs: list[float] = []

with open("./bank_mandiri/ATM_MANDIRI_OPERASIONAL_CLUSTER_SUMUT.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        results = get_lat_lng(row[0])
        addresses.append(row[0])
        lats.append(results[1]['lat'])
        lngs.append(results[1]['lng'])
```

```python
len(addresses), len(lats), len(lngs)
```

```python
pd.DataFrame(data={'address': addresses, 'lat': lats, 'lng': lngs}).to_csv('./bank_mandiri/lat_lng_ATM_MANDIRI_OPERASIONAL_CLUSTER_SUMUT.csv', index=False)
```
