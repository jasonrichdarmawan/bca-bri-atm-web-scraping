```python
load_dotenv()

GEOCODING_API_KEY = os.getenv('GEOCODING_API_KEY')
```

```python
def get_lat_lng(address: str):
    r = requests.get(f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={GEOCODING_API_KEY}')
    results = r.json()['results'][0]
    return results['formatted_address'], results['geometry']['location']
```

```python
get_lat_lng(address)
```

```python
addresses: list[str] = []
lats: list[float] = []
lngs: list[float] = []
cities: list[str] = []
provinces: list[str] = []

with open("./bank_mandiri/csv/ATM_MANDIRI_OPERASIONAL_CLUSTER_SUMUT.csv", 'r') as file:
    csvreader = csv.reader(file)
    for row in csvreader:
        results = get_lat_lng(row[0])
        addresses.append(row[0])
        cities.append(row[1])
        provinces.append(row[2])
        lats.append(results[1]['lat'])
        lngs.append(results[1]['lng'])
```

# Automated

This part is automated.

```python
def save_lat_long(filepath: str):
    addresses: list[str] = []
    lats: list[float] = []
    lngs: list[float] = []
    cities: list[str] = []
    provinces: list[str] = []
    
    with open(f'./bank_mandiri/csv/{filepath}', 'r') as file:
        csvreader = csv.reader(file)
        for row in csvreader:
            try:
                results = get_lat_lng(row[0])
            except IndexError:
                print('can\'t get coordinate of', row[0])
                continue
            
            addresses.append(row[0])
            cities.append(row[1])
            provinces.append(row[2])
            lats.append(results[1]['lat'])
            lngs.append(results[1]['lng'])
            
    pd.DataFrame(data={'address': addresses, 'city': cities, 'province': provinces, 'latitude': lats, 'longitude': lngs}).to_csv(f'./bank_mandiri/lat_long/{filepath}', index=False)
```
